```c++
// Functions in C++, Part 2

// (2) A recursive fibonacci sequence funtion, which calculates the nth fibonacci number.
//     NOTE: f0 = 0, f1 = 1, fn = f_{n-1} + f_{n-2}

// BEGIN (2)

int fib(int n) {
    if (n == 0) {
        return 0;
    } else if (n == 1) {
        return 1;
    } else {
        int f_n1 = fib(n - 1);
        int f_n2 = fib(n - 2);
        return f_n1 + f_n2;
    }
}

// END (2)
```

```c++
// Functions in C++, Part 3

// (3) A main function which calculates the 0th through the 7th Fibonacci numbers.
//     Use the fib() function!

// BEGIN (3)

int main() {
    
    for (int i = 0; i <= 7; i++) {
        std::cout << i << ": " << fib(i) << std::endl;
    }
    
    return 0;
}

// END (3)

// In a normal program, you wouldn't need to call the main function.
// However, since this is a Jupyter notebook, you have to call it manually
main();
```

```c++
// Arrays

// (6) Complete the function betterFib. Use an array instead of recursion!
//     * If n is 0 or 1, return the answer
//     * Otherwise, declare an array fibArr.
//         Determine the ith Fibonacci number using fibArr[i-1] and fibArr[i-2]
//         The last element in fibArr should be the answer!
//     NOTE: f0 = 0, f1 = 1, fn = f_{n-1} + f_{n-2}

int betterFib(int n);
```

```c++
int betterFib(int n) {
    // BEGIN (6)
    
    if (n == 0 or n == 1) 
        return n;
    else {
        int fibArr[n+1];
        fibArr[0] = 0;
        fibArr[1] = 1;
        
        for (int i = 2; i <= n; i++)
            fibArr[i] = fibArr[i-1] + fibArr[i-2];
        
        return fibArr[n];
    }
    
    // END (6)
}
```

```c++
// Structs

// Structs are a good way to organize related information!
// (7) Write a struct called "StudentInfo" which contains the following info:
//       * char array of size 30 called "name"
//       * uint8_t called "age"
//       * bool called "isGoodStudent"
// HINT: use "typedef" so you can use "StudentInfo" instead of "struct StudentInfo"

// BEGIN (7)

typedef struct{
    char name[30];
    int age;
    bool isGoodStudent;
} StudentInfo;

// END (7)

void allAboutYou(StudentInfo info);
```
